# 分类:暴力
## 思路1 从暴力算法开始


````python

import sys
import time
def solve1(n):
    ans = get_shortest(n)

def get_shortest(n):
    if n <= 1:#实际不会遇到求0的最短序列，该情况已被包含在i**2==n中;而1的最短序列不能在range中求解，所以单独写在前面;
        return 1
    res = sys.maxsize
    for i in range(1,n):#从1开始，逐步计算剩余值(n-i**2)，并求剩余值的最短序列长度
        if i**2 == n:#如果当前待求解的值已是一个平方数，则最短序列为自身，长度=1;
            return 1
        if i**2 >n:#剩余值一定是正数
            break
        rem = n - i**2
        ans = get_shortest(rem) + 1 # get_shortest()返回第一个子树的根节点的最短序列长度; +1 是序列中加入当前子树根节点与get_shortest()之间边的数字
        if ans < res:#记录当前子树根节点的最短序列长度
            res = ans
    return res
st=time.time()    
n=30
ans = get_shortest(n)
print('solve {}, ans {}'.format(n,ans))
print('consume:{}'.format(time.time()-st))
````

# 分类:暴力
## 思路2 加入备忘录机制

````python

import sys
import time
memo=dict()
def solve1(n):
    ans = get_shortest(n)

def get_shortest(n):
    global memo
    ans = memo.get(n,-1)
    if ans!=-1:#如果之前计算过
        return ans
    else:#如果之前没计算过，则现在开始计算
        res = sys.maxsize
        for i in range(1,n):#从1开始，逐步计算剩余值(n-i**2)，并求剩余值的最短序列长度
            if i**2 == n:#如果当前待求解的值已是一个平方数，则最短序列为自身，长度=1;
                memo[n] = 1
                return memo[n]
            if i**2 >n:#剩余值一定是正数
                break
            rem = n - i**2
            ans = get_shortest(rem) + 1 # get_shortest()返回第一个子树的根节点的最短序列长度; +1 是序列中加入当前子树根节点与get_shortest()之间边的数字
            if ans < res:#记录当前子树根节点的最短序列长度
                res = ans
        memo[n]=res
        return memo[n]
st=time.time()
n=30
solve1(n)
print('solve {}, ans {}'.format(n,memo[n]))
print('consume:{}'.format(time.time()-st))
````

# 分类：动态规划
## 思路3 自底向上

````python
import sys
import time
def solve(n):
    ans = dict()
    ans[1]=1
    for i in range(2,n+1):
        val = ans.get(i,sys.maxsize)#若没计算过当前值，则取无穷大
        for j in range(1,i):
            # val = ans.get()
            if j**2 == i:#如果是完全平方数则值=1,且计算数组中下一个值
                val = 1
                break
            if j**2 > i:
                break
            #TODO : 计算res_temp
            rem = i-j**2
            res_temp = ans[rem]+1
            
            if res_temp< val:
                val = res_temp

        ans[i] = val
    
    return ans[n],ans
        




st=time.time()
n=8
ans,dic = solve(n)
print('solve {}, ans {}'.format(n,ans))
print(dic)
print('consume:{}'.format(time.time()-st))
````

# 分类：动态规划
## 思路3 自底向上
### 改进：平方数跳过
#### 分析：上面一种和现在这种方法，即使自底向上的DP,但仍超时，时间复杂度nlogn


````python
import sys
import time
import math
def isSquare(num):
    sqrt_ = int(math.sqrt(num))
    if sqrt_**2 == num:
        return True
    else:
        return False
def solve(n):
    ans = dict()
    for i in range(1,n+1):
        if isSquare(i):# NOTE 改进1 平方数可直接赋1
            ans[i] = 1
            continue

        for j in range(1,i):
            if j**2 > i:
                break
            #TODO : 计算res_temp
            rem = i-j**2
            res = ans[rem]+1
            
            if res <= ans.get(i,i):
                ans[i] = res
    
    return ans[n],ans
        




st=time.time()
n=8
ans,dic = solve(n)
print('solve {}, ans {}'.format(n,ans))
print(dic)
print('consume:{}'.format(time.time()-st))
````

class Solution {

        int sqrt_;

        if (sqrt_*sqrt_ == num){
            return 1;
        }

    int numSquares(int n) {

                ans[i] = 1;

                }
            }
        }

实际上这道题可以用BFS来解，求解值视为图的源节点，目标值0视为图的终节点，那么我们试图在图中找到从源到终的最短路径，得到其长度。在这样的模型中，平方数是边，状态数是节点，所以如果两个数之间相差一个平方数，则有一条边相连。那么这种模型就类似岛屿数量/转盘锁的题目，每个节点的邻节点是若干种状态。
例如从源12到目标0:
* 可以由12经过边9到3，再由3经过边1到2，依次到0;
* 也可以从12由边4到8，再由边4到4，再由边4到0.

</br>
第一种方法经过边9,1,1,1，路径长度4;第二种方法经过边4,4,4，路径长度3.
而经过的边就是我们求到的组合平方数.

````python
#超时
import sys
import time
from math import sqrt
from queue import Queue

def isSquare(num):
    i = int(sqrt(num))
    if i**2 == num:
        return True
    else:
        return False
def solve1(n):
    q = Queue()
    ans = get_shortest(n,q)
    return ans

def get_shortest(n,q):
    step = 0
    stat = []
    q.put((n,step))
    while not q.empty():
        unit = q.get()
        increment(unit,q,stat)
    return stat[0]
    
def increment(unit,q,stat):
    num, step = unit
    step += 1 
    if isSquare(num) :#如果是平方数
        if len(stat)>0 : #要不能获得更短序列
            if step < stat[0]:
                stat.append(step)
        elif len(stat)==0:#，要不还没找到更短序列
            stat.append(step)

    
    elif len(stat)==0: #或是还未找到最短序列长度
        
        for i in range(num):
            if i**2 > num:
                break
            q.put((num-i**2,step))
    elif step<stat[0]:#如果增加一步,小于最短序列长度;
        for i in range(num):
            if i**2 > num:
                break
            q.put((num-i**2,step))
    

st=time.time()    
n=7168
ans = solve1(n)
print('solve {}, ans {}'.format(n,ans))
print('consume:{}'.format(time.time()-st))
````

在上一个方法中，设置stat来记录最短序列长度是多余的，因为bfs本身的visited机制就可以知道，已访问过的节点再下一次访问时，长度肯定是增加了。所以在计算新的节点时，只要出现0就可得到最短路径

````python
#AC
import sys
import time
from math import sqrt
from queue import Queue

def isSquare(num):
    i = int(sqrt(num))
    if i**2 == num:
        return True
    else:
        return False
def solve1(n):
    q = Queue()
    ans = get_shortest(n,q)
    return ans

def get_shortest(n,q):
    step = 0
    # stat = []
    q.put((n,step))
    visited = set()
    while not q.empty():
        num, step = q.get()
        step += 1#开始找邻节点的时候step+1
        rem = []
        for i in range(1,int(num**0.5)+1):
            if num-i**2 == 0:
                return step
            rem.append(num - i**2)
        for j in rem:
            if j not in visited:
                q.put((j,step))
                visited.add(j)


st=time.time()    
n=7168
ans = solve1(n)
print('solve {}, ans {}'.format(n,ans))
print('consume:{}'.format(time.time()-st))
````
