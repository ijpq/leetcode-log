# 分类:暴力
## 思路1 从暴力算法开始


````python

import sys
import time
def solve1(n):
    ans = get_shortest(n)

def get_shortest(n):
    if n <= 1:#实际不会遇到求0的最短序列，该情况已被包含在i**2==n中;而1的最短序列不能在range中求解，所以单独写在前面;
        return 1
    res = sys.maxsize
    for i in range(1,n):#从1开始，逐步计算剩余值(n-i**2)，并求剩余值的最短序列长度
        if i**2 == n:#如果当前待求解的值已是一个平方数，则最短序列为自身，长度=1;
            return 1
        if i**2 >n:#剩余值一定是正数
            break
        rem = n - i**2
        ans = get_shortest(rem) + 1 # get_shortest()返回第一个子树的根节点的最短序列长度; +1 是序列中加入当前子树根节点与get_shortest()之间边的数字
        if ans < res:#记录当前子树根节点的最短序列长度
            res = ans
    return res
st=time.time()    
n=30
ans = get_shortest(n)
print('solve {}, ans {}'.format(n,ans))
print('consume:{}'.format(time.time()-st))
````

# 分类:暴力
## 思路2 加入备忘录机制

````python

import sys
import time
memo=dict()
def solve1(n):
    ans = get_shortest(n)

def get_shortest(n):
    global memo
    ans = memo.get(n,-1)
    if ans!=-1:#如果之前计算过
        return ans
    else:#如果之前没计算过，则现在开始计算
        res = sys.maxsize
        for i in range(1,n):#从1开始，逐步计算剩余值(n-i**2)，并求剩余值的最短序列长度
            if i**2 == n:#如果当前待求解的值已是一个平方数，则最短序列为自身，长度=1;
                memo[n] = 1
                return memo[n]
            if i**2 >n:#剩余值一定是正数
                break
            rem = n - i**2
            ans = get_shortest(rem) + 1 # get_shortest()返回第一个子树的根节点的最短序列长度; +1 是序列中加入当前子树根节点与get_shortest()之间边的数字
            if ans < res:#记录当前子树根节点的最短序列长度
                res = ans
        memo[n]=res
        return memo[n]
st=time.time()
n=30
solve1(n)
print('solve {}, ans {}'.format(n,memo[n]))
print('consume:{}'.format(time.time()-st))
````

# 分类：动态规划
## 思路3 自底向上

````python
import sys
import time
def solve(n):
    ans = dict()
    ans[1]=1
    for i in range(2,n+1):
        val = ans.get(i,sys.maxsize)#若没计算过当前值，则取无穷大
        for j in range(1,i):
            # val = ans.get()
            if j**2 == i:#如果是完全平方数则值=1,且计算数组中下一个值
                val = 1
                break
            if j**2 > i:
                break
            #TODO : 计算res_temp
            rem = i-j**2
            res_temp = ans[rem]+1
            
            if res_temp< val:
                val = res_temp

        ans[i] = val
    
    return ans[n],ans
        




st=time.time()
n=8
ans,dic = solve(n)
print('solve {}, ans {}'.format(n,ans))
print(dic)
print('consume:{}'.format(time.time()-st))
````

# 分类：动态规划
## 思路3 自底向上
### 改进：平方数跳过
#### 分析：上面一种和现在这种方法，即使自底向上的DP,但仍超时，时间复杂度nlogn


````python
import sys
import time
import math
def isSquare(num):
    sqrt_ = int(math.sqrt(num))
    if sqrt_**2 == num:
        return True
    else:
        return False
def solve(n):
    ans = dict()
    for i in range(1,n+1):
        if isSquare(i):# NOTE 改进1 平方数可直接赋1
            ans[i] = 1
            continue

        for j in range(1,i):
            if j**2 > i:
                break
            #TODO : 计算res_temp
            rem = i-j**2
            res = ans[rem]+1
            
            if res <= ans.get(i,i):
                ans[i] = res
    
    return ans[n],ans
        




st=time.time()
n=8
ans,dic = solve(n)
print('solve {}, ans {}'.format(n,ans))
print(dic)
print('consume:{}'.format(time.time()-st))
````

```python
class Solution {
public:
    int isSquare(float num){
        int sqrt_;
        sqrt_ = int(sqrt(num));
        if (sqrt_*sqrt_ == num){
            return 1;
        }
        return 0;
    }
    
    int numSquares(int n) {
        int ans[n+1] ;
        for (int i = 1; i != n+1; i++){
            ans[i] = i;
            if (isSquare(i)){
                ans[i] = 1;
                continue;
            }
                
            for (int j = 1; j*j <= i; j++){
                int rem = i - j*j ;
                int res = ans[rem] + 1;
                if (res <= ans[i]){
                    ans[i] = res;
                }
            }
        }
        return ans[n];
    }
};
```
